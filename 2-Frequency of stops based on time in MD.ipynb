# 2-Which hours of a day got most stops in Maryland?

##Assumptions: 6-12 morning    12-18 afternon    18-24 night    24-6 mid night

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("E:/Education/Auburn/2018-Fall/Info-Jeff/project/Traffic_Violations.csv")
```

```python
df2=df[['TimeOfStop']]
df2
```

```python
type(df2.loc[2])
```

```python
df['TimeOfStop']=pd.to_datetime(df2.TimeOfStop)
```

```python
df3=df[['TimeOfStop']]
df4=df.TimeOfStop.dt.hour
df4
```

```python
df8=df.loc[(df['State'] == 'MD')]
```

```python
df5=df8[['TimeOfStop']]
df5
```

```python
###Stops which are occured at morning
mask1 = ((df4>= 6) & (df4< 12))
       
M=df5.loc[mask1]
Mor=M.count()/df5.count()
Mor
```

```python
###Stops which are occured at afternoon
mask1 = ((df4>= 12) & (df4< 18))
       
A=df5.loc[mask1]
Af=A.count()/df5.count()
Af
```

```python
###Stops which are occured at night
mask1 = ((df4>=18) & (df4< 24))
       
N=df5.loc[mask1]
Ng=N.count()/df5.count()
Ng
```

```python
###Stops which are occured at midnight
mask1 = ((df4>=0) & (df4<6))
       
D=df5.loc[mask1]
Dt=D.count()/df5.count()
Dt
```

```python
import matplotlib.pyplot as plt
 
values = [Mor, Af, Ng,Dt]
colors = ['b', 'g', 'r', 'c']
labels = ['Morning', 'Afternoon', 'Night','Midnight']
explode = (0, 0, 0, 0)
plt.pie(values, colors=colors, labels=labels,
explode=explode, autopct='%1.1f%%',
counterclock=False, shadow=True)
plt.title('Frequency of stops based on time in MD')
plt.show()
```

# Conclusion

Based on pie chart we can say that lowest percentage of stops is for morning with 24.6% .
Also, 25% of stops are occurred during afternoon. But,most of the stops in Maryland are occurred during night.It is reasonable because during night it is more possible for drivers to do not see signs. Even the number of drivers after midnight is supposed to be very low but we have good percentage of stops, it is 19%.
